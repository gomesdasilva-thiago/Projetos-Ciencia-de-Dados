# Realizando coleta de dados em arquivos csv utilizando linguagem python;

```python
# Importando os pacotes necessários
import pandas as pd
import os
```

```python
# Definindo o caminho os estão os arquivos
DIRETORIO = 'C:\\Users\\thiag\\Documents\\Bootcamp Online IGTI\\Módulo 2\\Material Complementar\\Trabalho Prático\\base_trabalho_pratico'
DIRETORIO
```

```python
# Criando uma lista com o nome de todos os arquivos que estão no DIRETORIO
lista_df = []
lista_path = os.listdir(DIRETORIO)
lista_path[:5]
```

```python
# Abrindo cada arquivo e coletando os dados
lista_df = []
for arquivo in lista_path:
    local_arquivo = os.path.join(DIRETORIO, arquivo)
    df = pd.read_csv(local_arquivo, sep = "|", encoding='latin-1')
    lista_df.append(df)
df_pref = pd.concat(lista_df)
```

```python
# Renomeando a coluna data_coleta
df_pref = df_pref.rename(columns={'data_coleta' : 'data_pesquisa'})
```

```python
df_pessoas = df_pref[['cod_pessoa', 'genero', 'data_nascimento']].drop_duplicates()
df_animal_estimacao = list(df_pref.animal_estimacao.unique())
df_bebida = list(df_pref.bebida_favorita.unique())
df_hobbie = list(df_pref.hobbies.unique())
df_clima = list(df_pref.clima.unique())
```

# Inserindo os dados no banco de dados MySQL

```python
# Inserindo os dados na tabela pessoa
for pessoa in df_pessoas.itertuples():
    cod_pessoa = pessoa.cod_pessoa
    genero = pessoa.genero
    data_nascimento = str(pessoa.data_nascimento)
    try:
        query = f"""insert into pessoa (cod_pessoa, genero, data_nascimento ) 
                     values ('{cod_pessoa}', '{genero}', '{data_nascimento}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {cod_pessoa}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {cod_pessoa}. O erro foi encontrado foi: {e}")
```

```python
# Inserindo os dados na tabela animal_estimacao
for animal in df_animal_estimacao:
    try:
        query = f"""insert into animal_estimacao (animal) 
                     values ('{animal}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {animal}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {animal}. O erro foi encontrado foi: {e}")
```

```python
# Criando uma tabela staging no banco SQL para popular a tabela pesquisa através do MySQL
df_pref.to_sql('stg_pesquisa', con=conn, schema='trabalho_praticoi')
```

```python
# Populando a tabela pesquisa através de linguagem SQL no Python

query = f"""insert into pesquisa (data_pesquisa, cod_pessoa, cod_animal_estimacao, cod_clima, cod_bebida, cod_hobbie) 
                (
                select
                    stg.data_pesquisa,
                    p.cod_pessoa,
                    ae.cod_animal_estimacao,
                    c.cod_clima,
                    b.cod_bebida,
                    h.cod_hobbie              
                from stg_pesquisa stg
                    inner join animal_estimacao ae
                    on stg.animal_estimacao = ae.animal
                    inner join bebida b
                    on stg.bebida_favorita = b.bebida
                    inner join clima c
                    on stg.clima = c.clima
                    inner join hobbie h
                    on stg.hobbies = h.hobbie
                    inner join pessoa p
                    on stg.cod_pessoa = p.cod_pessoa              
                )
                      
            """      
conn.execute(query)      
        
```
